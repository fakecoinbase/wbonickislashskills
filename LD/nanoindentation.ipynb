## This notebook was created to load results of nanoindentation experiment and visualize it

### Nanoindentation is an indentation hardness test applied to small volumes. The results (in txt format) containts 19 measurement, each measuremnt consists of load and displacement values

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
#loading the data from txt file
res=[]
res_temp=[]
with open('results.txt','r') as f:   
    for line in f:
        if line != '\n':
            res_temp.append(line)
        if line=='\n':
            res.append(res_temp)
            res_temp=[]
```

```python
#cleaning the list with the results
res_clean=[]
for i in res:
    if i!=[]:
        res_clean.append([tuple(j.replace("\n","").split("\t")) for j in i])
```

```python
#there is 19 measurements, here the list of lists with each measurement for two variables - load and displacement - is created
load_list=[]
dis_list=[]
for i in res_clean:
    load_temp=[]
    dis_temp=[]
    for j in i:
#negative values are removed
        if (float(j[0])>0 and float(j[1])>0):
            load_temp.append(float(j[0]))
            dis_temp.append(float(j[1]))
    load_list.append(load_temp)
    dis_list.append(dis_temp)
```

```python
#printing the first measurement
for i,j in zip(load_list[0],dis_list[0]):
    print(i,j)
```

```python
#number of measurements
num_of_measurements=len(load_list)
#list for figure titles
titles=[]
for i in range(num_of_measurements):
    titles.append('{} experiment'.format(i+1))
```

```python
#plotting all data in one figure
sns.set(style="whitegrid", color_codes=True)

fig,ax = plt.subplots(figsize=(10,10))

for i,j,k in zip(dis_list,load_list,titles):
    ax.plot(i,j, label=k)

ax.set(xlim=[0,60], ylim=[0,1.1],
       xticks=np.arange(0,60,5),
       yticks=np.arange(0,1.2,0.1),
       xlabel='displacement [nm]', ylabel='load [mN]')
ax.set_title('nanoindentation load-displacement curves', fontsize=20)
plt.legend()
plt.show()
```

```python
#ploting ever curve in separate figure
for i,j,k in zip(dis_list,load_list,titles):
    plt.figure(figsize=(8,4))
    plt.plot(i,j)
    plt.xlim(0,60)
    plt.ylim(0,1.1)
    plt.xlabel('displacement $[nm]$ ',fontsize=15)
    plt.ylabel('load $[mN]$',fontsize=15)
    plt.xticks=(np.arange(0,60,5))
    plt.yticks=(np.arange(0,1.2,0.1))
    plt.title(k,fontsize=20)
    plt.show()
        
```
